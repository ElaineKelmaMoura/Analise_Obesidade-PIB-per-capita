# Projeto de obesidade entre adultos nos anos de 1965 e 2016.


* Limpe os dados do DataFrame, criando uma coluna de nome  'Obesity' que conterá os valores de obesidade. Transforme em float as colunas que porventura foram importadas como strings.


```
import pandas as pd
import numpy as np
```

```
df_obesity = pd.read_csv("obesity_cleaned.csv") #Leitura do arquivo
```

```
df_obesity
```

```
#Verificando as colunas
df_obesity["Year"].head(5)
```

```
df_obesity["Obesity (%)"].head(5)
```

```
df_obesity["Obesity (%)"].iloc[0]
```

```
df_obesity["Obesity (%)"].value_counts()
```

```
df_obesity.columns
```

```
#Deletando coluna
del df_obesity['Unnamed: 0']
```

```
#Criando uma nova coluna, e uma função para jogar fora informações desnecessárias.
```

```
df_obesity["Obesity (%)"]
```

```
df_obesity["Obesity"] = df_obesity["Obesity (%)"].apply(lambda x: x.split()[0])
```

```
df_obesity['Obesity'].head(5)#verificando o tipo da coluna
```

```
df_obesity['Obesity'].value_counts()#Necessário tirar o No
```

```
df_obesity.loc[df_obesity["Obesity"] == "No", "Obesity"] = np.nan
```

```
df_obesity["Obesity"] = df_obesity["Obesity"].dropna()
```

```
df_obesity["Obesity"].value_counts()
```

```
#Transformando para float
df_obesity["Obesity"] = df_obesity["Obesity"].apply(lambda x: float(x))
```

```
#Transformando para int
df_obesity["Year"] = df_obesity["Year"].apply(lambda x: int(x))
```

```
df_obesity
```

```
df_obesity.set_index("Year", inplace=True)#Defenindo o index - Ano
```

```
df_obesity
```



*  Qual o percentual médio de obesidade por sexo no mundo no ano de 2015?



```
df_obesity[df_obesity.index == 2015].groupby("Sex").mean(numeric_only=True)
```



*   Quais são os 5 países com a maior e menor taxa de aumento nos índices de obesidade no período observado?




```
df_obesity_start = df_obesity[df_obesity.index == 1975]
df_obesity_end = df_obesity[df_obesity.index == 2016]
```

```
df_obesity_start.head(5)
```

```
#Colocando como índice de ambos o país
df_obesity_start.set_index("Country", inplace=True)
df_obesity_end.set_index("Country", inplace=True)
```

```
#Obsidade de todos os países no ano final por país
df_obesity_end[df_obesity_end["Sex"] == "Both sexes"]["Obesity"]
```

```
#fazendo analise com ambos os sexos
df_obesity_evol = df_obesity_end[df_obesity_end["Sex"] == "Both sexes"]["Obesity"] - df_obesity_start[df_obesity_start["Sex"] == "Both sexes"]["Obesity"]
```

```
df_obesity_evol #Evoluções percentuais no período observado de todos os países.
```

```
df_obesity_evol.sort_values() #Ordenando
```

```
df_obesity_evol.sort_values().dropna().head(5) #Jogando fora países com informações NaN / 5 países com menor evolução
```

```
df_obesity_evol.sort_values().dropna().tail(5) #Países com maior evolução
```

```
df_obesity_end[df_obesity_end.index == "Tuvalu"]
```



*   Quais os países com maiores e menores níveis percentuais de obesidade em 2015?
